```python
from rdkit import Chem
PeriodicTable = Chem.GetPeriodicTable()
```

```python


simbolo_elementos = []
for i in range(1,21):
   simbolo_elementos.append(PeriodicTable.GetElementSymbol(i))
# Ponemos las variables que tendrá nuestra tabla
numero_atomico = [PeriodicTable.GetAtomicNumber(x) for x in simbolo_elementos]
masa_atomica = [PeriodicTable.GetAtomicWeight(x) for x in simbolo_elementos]
nombre_del_elemento = [PeriodicTable.GetElementName(x) for x in range(1,21)]
valencias =  [PeriodicTable.GetValenceList(x) for x in range(1,21)]


import pandas as pd
Datos = {'Nombre': nombre_del_elemento, 'Símbolo': simbolo_elementos, 'Masa atómica': masa_atomica, 'Número atómico': numero_atomico,'valencias': valencias}
df = pd.DataFrame(Datos)
df
```

## Pregunta 2: Configuración electrónica (4 puntos)

Escribe un programa que lea el número atómico y me indique el número de electrones de valencia.

```python
# Número de protones como input

def electrones_de_valencia(numero_atomico):
    # Diccionario que contiene el numero atomico con su respecctivo numero de valencia
    tabla_numero_atomico_valencia = {1:1, 3:1, 11:1, 19:1, 37:1, 55:1, 87:1, 4:2, 12:2, 20:2, 38:2, 56:2, 88:2, 5:3, 13:3, 31:3, 49:3, 81:3, 113:3, 6:4, 14:4, 32:4, 50:4, 82:4, 114:4, 7:5, 15:5, 33:5, 51:5, 83:5, 115:5, 8:6, 16:6, 34:6, 52:6, 84:6, 116:6, 9:7, 17:7, 35:7, 53:7, 85:7, 117:7, 2:8, 10:8, 18:8, 36:8, 54:8, 86:8, 118:8  }

    # Condicional para verificar que el numero aatomico ingresado esta en mi diccionario
    if numero_atomico in tabla_numero_atomico_valencia:
        return tabla_numero_atomico_valencia[numero_atomico]
    else:
        return "Número atómico no válido."

# Solicio al usuario que ingrese el número atómico que desee
try:
    numero_atomico = int(input("Ingresa el número atómico del elemento: "))
    resultado = electrones_de_valencia(numero_atomico)
    if resultado != "Número atómico no válido.":
        print(f"El número de electrones de valencia para el número atómico {numero_atomico} es {resultado}.")
    else:
        print(resultado)
except ValueError:
    print("Por favor, ingrese un número atómico válido.")
```
